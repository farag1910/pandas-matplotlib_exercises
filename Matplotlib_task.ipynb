```python
import pandas as pd
import matplotlib.pyplot as plt

financial_data = [
    {
        "Date": "10-03-16",
        "Open": 774.25,
        "High": 776.065002,
        "Low": 769.5,
        "Close": 772.559998
    },
    {
        "Date": "10-04-16",
        "Open": 776.030029,
        "High": 778.710022,
        "Low": 772.890015,
        "Close": 776.429993
    },
    {
        "Date": "10-05-16",
        "Open": 779.309998,
        "High": 782.070007,
        "Low": 775.650024,
        "Close": 776.469971
    },
    {
        "Date": "10-06-16",
        "Open": 779.0,
        "High": 780.47998,
        "Low": 775.539978,
        "Close": 776.859985
    },
    {
        "Date": "10-07-16",
        "Open": 779.659973,
        "High": 779.659973,
        "Low": 770.75,
        "Close": 775.080017
    }
]

data=pd.DataFrame(financial_data)

plt.plot(data['Date'], data['Open'], label='Open', marker='o')

plt.plot(data['Date'], data['High'], label='High', marker='o')

plt.plot(data['Date'], data['Low'], label='Low', marker='o')

plt.plot(data['Date'], data['Close'], label='Close', marker='o')

plt.title('Alphabet Inc. Financial Data (Oct 3, 2016 - Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(data['Date'], data['Close'], marker='o', linestyle='-', color='red', label='Close')
plt.grid(True, linestyle='-', linewidth=0.5, color='blue')
```

```python
data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 4],
    'c': [8, 3, 7, 6, 4],
    'd': [5, 4, 4, 4, 3],
    'e': [7, 2, 7, 8, 3],
    'f': [6, 6, 8, 6, 2]
}

df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])

df.plot(kind='bar', figsize=(10, 6))

plt.title('Bar Plot from DataFrame')
plt.xlabel('Index')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

```python
import numpy as np 
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['A', 'B', 'C', 'D']
bars = plt.bar(labels, mean_velocity, yerr=std_deviation, color='skyblue', capsize=5)
plt.title('Bar Plot with Error Bars and Mean Values')
plt.xlabel('Groups')
plt.ylabel('Mean Velocity')
plt.tight_layout()
plt.show()
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
plt.pie(popularity,labels=languages)
```
